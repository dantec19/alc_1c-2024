Ejercicio 1. Hacer un programa que reciba una matriz A ∈ Rn×n y un entero positivo k y que aplique k
iteraciones del m´etodo de la potencia con un vector aleatorio inicial v ∈ Rn. El programa debe devolver
un vector a ∈ Rk, donde ai sea la aproximaci´on al autovalor obtenida en el paso i.

```python
def ast(A):
    return np.transpose(np.conjugate(A))
```

```python
def metodo_potencia(A, k):
    v = np.random.rand(k, 1)
    avals_aprox = np.zeros(np.shape(A)[0])
    for i in range(k):
        w = A @ v
        aval_aprox = np.dot(ast(v), w) / np.dot(ast(v), v)
        avals_aprox[i] = aval_aprox
        v = w / np.linalg.norm(w)
        
    return avals_aprox
```
